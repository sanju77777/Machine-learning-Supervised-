```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
dummies = pd.get_dummies(df.town)
dummies = dummies * 1
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['town','robinsville'], axis=1)
final
```

```python
x = final.drop(['price'], axis=1)
x
```

```python
y = final.price
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict(x)
```

```python
model.predict([[3500,1,0]])
```

```python
model.score(x,y)
```

```python
model.predict([[2452,0,1]])
```
